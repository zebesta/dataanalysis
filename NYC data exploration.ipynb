```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

filename = 'nyc_subway_weather.csv'
subway_df = pd.read_csv(filename)
def hourly(series):
    hourly = series - series.shift(1)
    return hourly
print subway_df.head()
# print subway_df.groupby('UNIT')[['ENTRIESn', 'EXITSn']].apply(hourly)['ENTRIESn'].plot()
grouped_and_shifted = subway_df.groupby('UNIT')[['ENTRIESn', 'EXITSn']].apply(hourly)
subway_df.groupby('UNIT')['ENTRIESn_hourly'].max().plot()
print grouped_and_shifted.head()
```
